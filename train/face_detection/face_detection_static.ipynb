### Face detaction of a static image

Based on datacamp tutorial "Face Detection with Python using OpenCV": https://www.datacamp.com/tutorial/face-detection-python-opencv

```python
# pip install opencv-python # Run once to install the library
import cv2 # OpenCV
```

### Function `detect_faces` accepts path to an image and returns an image with a drawn bounding box around all detected faces

```python
def detect_faces(path_to_image):
    img = cv2.imread(path_to_image)
    gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    face_classifier = cv2.CascadeClassifier(
        cv2.data.haarcascades + "haarcascade_frontalface_default.xml" # pretrained model
    )
    faces = face_classifier.detectMultiScale(
        gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(40, 40)
    )
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 4)
    return {
        'result': cv2.cvtColor(img, cv2.COLOR_BGR2RGB),
        'faces': faces
    }
    
```

### Testing of `detect_faces`. Please, replace imagePath with a path to a test image

```python
# Test
imagePath = 'input_image.jpg'
output = detect_faces(imagePath)

# Render output image
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plt.imshow(output['result'])
plt.axis('off')

output['faces']
```
